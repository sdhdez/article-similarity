```python
kleis = kl.load_corpus()
kleis.training(features_method="simple-posseq", filter_min_count=16)
```

### [Aminer Citation Network Datasets - v1 and v9](https://aminer.org/citation)



```python
#data_path = "resources/aminer/v1"
data_path = "../../src/somhos/resources/aminer/v9"
```

### Select related documents or not

```python
related_docs = True
```

```python
# Load document ids in the Same order than the matrices
indexdocs = rd.get_sample_ids(data_path, related_docs=related_docs)
# Index 
ix_data = rq.cur_indexed_docs(data_path)
```

```python
for idoc in indexdocs[:100]:
    result = rq.find_indexdoc(ix_data, idoc)
    for r in result:
        text = r['content']
        keyphrases = kleis.label_text(text, post_processing=False)
        print(text)
        print("\n  ++ ", set([kp[2] for kp in keyphrases]))
        print("\n----\n")
```
